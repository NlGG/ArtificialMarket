```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def coupon_bonds():
    F = 100
    T = 5
    C = 10
    r = 0.12
    for i in range(10):
        y = []
        x = []
        v = 0
        i = float(i)
        t = i/2.0
        t_var = 0
        while t <= 5:
            if t in [1.0, 2.0, 3.0, 4.0] and t_var != 0:
                v += 10.0*math.exp(-t_var*r)
                x.append(t)
                y.append(v)
            if t == 5.0:
                v += 110.0*math.exp(-(t_var*r))
                x.append(t)
                y.append(v)
            t += 0.5
            t_var += 0.5
        plt.plot(x,y)
```

```python
coupon_bonds()
```

```python
def risk_neutral():
    u = 0.2
    r = 0.1
    d = -1
    x = []
    y = []
    U = int(u*500+500)
    for d in range(U-1):
        d = float(d)
        d = (d-499.0)/500.0
        p = (r-d)/(u-d)
        if p < 0:
            break
        x.append(d)
        y.append(p)
    plt.plot(x,y)
```

```python
risk_neutral()
```
