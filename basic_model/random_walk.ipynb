```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
```

```python
def random_walk():
    x = np.array([0])
    for i in range(10000):
        x = np.append(x, x[i]+np.random.choice([1, -1]))
    return x
```

```python
x = random_walk()
plt.plot(x)
```

```python
def random_walk_2d():
    x = [0]
    y = [0]
    for i in range(1000):
        x = np.append(x, x[i]+np.random.choice([1, -1]))
        y = np.append(y, y[i]+np.random.choice([1, -1]))
    return x, y
```

```python
x = random_walk()
plt.plot(x)
```

```python
x, y = random_walk_2d()
plt.plot(x, y)
```

```python
def random_walk_3d():
    x = [0]
    y = [0]
    z = [0]
    for i in range(1000):
        x = np.append(x, x[i]+np.random.choice([1, -1]))
        y = np.append(y, y[i]+np.random.choice([1, -1]))
        z = np.append(z, z[i]+np.random.choice([1, -1]))
    return x, y, z
```

```python
x, y, z = random_walk_3d()
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter3D(x, y, z)
```

```python
def return_0():
    return_0 = np.array([0])
    for i in range(100):
        x, y, z = random_walk_3d()
        for i in range(1,1000):
            if x[i] == 0 and y[i] == 0 and z[i] == 0:
                return_0 += 1
                break
    return return_0
return_0()
```

```python
np.average([return_0() for i in range(10)])
```
